```python
%matplotlib inline
```

```python
# Exempel av Stefan Blecko
def competence(**skills):
    x = skills.keys()
    y = skills.values()
    plt.ylim(0, 5)
    plt.bar(x, y)
    plt.ylabel('Kunskapsnivå')
    plt.xlabel('Certifierad i Microsoft MS-900 och SC-900.', 
               fontdict={'fontsize': 8}, labelpad=10)
    plt.show()

competence(Operativsystem=3, Programmering=2, Support=4.5, Molntjänster=3)
```
