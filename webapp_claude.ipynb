## 2. Vad klarar en Artifact tekniskt sett?

En React-Artifact körs i en sandlådad förhandsvisning direkt i webbläsaren. Allt som är ren klientlogik fungerar fullt ut:

- Villkor (if/else, switch, ternary-uttryck)
- Loopar (for, while, map/filter/reduce på arrayer)
- State och komponentlivscykel (useState, useEffect, useReducer)
- Bibliotek som recharts, lodash, d3, three.js, mathjs, papaparse med flera

Exempel – en enkel att-göra-lista med loop och villkor. Notebookens kernel är Python, så cellen nedan körs inte, men visas som en kodcell precis som exemplet i PDF:en:

```python
function TodoList() {
  const [items, setItems] = useState([]);
  const [text, setText] = useState("");

  function addItem() {
    if (text.trim() === "") return; // if-sats
    setItems([...items, { text, done: false }]);
    setText("");
  }

  return (
    <div>
      {items.map((item, i) => ( // loop
        <p key={i} style={{
          textDecoration: item.done ? "line-through" : "none"
        }}>{item.text}</p>
      ))}
    </div>
  );
}
```

## 3. Var gränsen går: databas, lagring och gratisversionen

Här skiljer sig gratisversionen från betalplanerna (Pro, Max, Team, Enterprise), enligt Anthropics officiella hjälpcenter:

- **Persistent lagring (window.storage)** – gör att en Artifact kan spara data mellan sessioner (t.ex. en journal eller en topplista). Detta kräver Pro/Max/Team/Enterprise, och artifacten måste dessutom publiceras för att lagringen ska fungera.
- **MCP-integrationer** – att koppla en Artifact direkt mot externa tjänster (t.ex. en riktig databas via en MCP-server) kräver också en betald plan.
- **Ingen egen backend** – oavsett plan körs Artifacts i webbläsarens sandlåda utan egen serverdel. En Artifact kan anropa externa API:er med fetch() om de tillåter CORS-anrop, men den kan inte prata direkt med t.ex. en SQL-databas – det kräver alltid ett API-lager emellan.
- **Ingen driftsättning/hosting** – Artifacts är en förhandsvisning, inte en publik webbserver. För en "riktig" app i produktion exporterar du koden till en egen utvecklingsmiljö och hostar den själv (Vercel, Netlify, GitHub Pages, egen server).

**Slutsats:** för snabb prototyping, UI-logik, kalkylatorer och visualiseringar räcker gratisversionen gott. Vill du ha en app som verkligen sparar data mellan besök eller pratar med en riktig databas, behöver du antingen en betald plan (persistent lagring/MCP) eller exportera koden och koppla en egen backend (t.ex. Supabase, Firebase eller ett eget REST-API).

## 4. Arbetsflödet steg för steg

### Steg 1 – Beskriv appen

Skriv i klartext vad appen ska göra, t.ex. "bygg en att-göra-lista med kategorier och filtrering". Ju tydligare krav (funktioner, dataflöde, känsla), desto bättre förstautkast.

### Steg 2 – Claude skapar en Artifact

Claude väljer lämpligt format (oftast React/JSX) och genererar en körbar komponent som visas direkt i sidopanelen – ingen installation, inget bygge-steg.

### Steg 3 – Iterera i konversationen

Be om ändringar i vanligt tal: "gör knappen större", "lägg till en till kategori", "validera formuläret". Claude uppdaterar samma Artifact istället för att börja om.

### Steg 4 – Testa direkt i förhandsgranskningen

Klicka runt i appen i panelen för att kontrollera att villkor, loopar och state beter sig som tänkt, innan du tar med koden vidare.

### Steg 5 – Exportera koden

Kopiera koden till din egen editor eller till ett GitHub-repo (t.ex. ett eget scriptlab-liknande repo) för vidare arbete i VS Code eller Codespaces.

### Steg 6 – Koppla en riktig backend/databas

Lägg till ett API-lager (eget REST-API, Supabase, Firebase) som Artifacten – eller den exporterade appen – pratar med via fetch(). På betalplan kan detta även ske via MCP direkt i Artifacten.

### Steg 7 – Driftsätt

Publicera den färdiga appen på en hostingtjänst (Vercel, Netlify, egen server) för att göra den tillgänglig för andra än dig själv.

Med andra ord: Claude Sonnet 5 och Artifacts täcker steg 1–4 mycket bra, redan i gratisversionen. Steg 5–7 är där en prototyp blir en riktig, driftsatt webbapplikation – och det sker till stor del utanför själva chatten, i din egen utvecklingsmiljö.
